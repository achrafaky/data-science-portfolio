```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Configuration du style visuel
sns.set_theme(style="darkgrid")

# Chemin dynamique adaptatif (Git-Ready)
if os.path.exists("04-Logistics-Operations/museum_visitors.csv"):
    file_path = "04-Logistics-Operations/museum_visitors.csv"
else:
    file_path = "museum_visitors.csv"

if os.path.exists(file_path):
    print(f"⏳ Chargement des données depuis : {file_path} ...")
    # Chargement avec la date en index pour activer les graphiques temporels (Time Series)
    data = pd.read_csv(file_path, index_col="Date", parse_dates=True)
    print(f"✅ Fichier chargé avec succès ! (Taille : {data.shape})\n")

    # ==========================================
    # 1. VUE D'ENSEMBLE : TOUS LES MUSÉES
    # ==========================================
    plt.figure(figsize=(16, 6))
    
    # Trace les lignes pour toutes les colonnes du DataFrame
    sns.lineplot(data=data, linewidth=2.5)
    
    plt.title("Évolution globale des visiteurs par musée au fil du temps", fontsize=16, pad=15)
    plt.xlabel("Année", fontsize=12)
    plt.ylabel("Nombre total de visiteurs", fontsize=12)
    
    # Sortie de la légende pour ne pas cacher les courbes
    plt.legend(title="Musées", bbox_to_anchor=(1.01, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # ==========================================
    # 2. COMPARAISON CIBLÉE : Avila Adobe vs Chinese American Museum
    # ==========================================
    plt.figure(figsize=(16, 6))
    
    # Sélection optimisée des deux musées pour une comparaison directe
    museums_to_compare = ['Avila Adobe', 'Chinese American Museum']
    sns.lineplot(data=data[museums_to_compare], linewidth=2.5)
    
    plt.title("Comparaison ciblée : Avila Adobe vs Chinese American Museum", fontsize=16, pad=15)
    plt.xlabel("Année", fontsize=12)
    plt.ylabel("Nombre de visiteurs", fontsize=12)
    plt.legend(title="Musées comparés")
    plt.tight_layout()
    plt.show()

else:
    print("❌ Erreur : Impossible de trouver le fichier CSV. Vérifie l'arborescence du projet.")
```
