```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Configuration du style visuel
sns.set_theme(style="darkgrid")

# Chemin dynamique adaptatif (Git-Ready)
if os.path.exists("05-Botanical-Research/iris.csv"):
    file_path = "05-Botanical-Research/iris.csv"
else:
    file_path = "iris.csv"

if os.path.exists(file_path):
    print(f"⏳ Chargement des données depuis : {file_path} ...")
    iris_data = pd.read_csv(file_path, index_col="Id")
    print(f"✅ Fichier chargé avec succès ! (Taille : {iris_data.shape})\n")

    # ==========================================
    # 1. DISTRIBUTION GLOBALE (KDE Plot)
    # ==========================================
    plt.figure(figsize=(12, 6))
    
    # Tracé des densités pour les 4 caractéristiques
    sns.kdeplot(data=iris_data, x='Petal Length (cm)', fill=True, alpha=0.5, label='Longueur Pétale')
    sns.kdeplot(data=iris_data, x='Sepal Length (cm)', fill=True, alpha=0.5, label='Longueur Sépale')
    sns.kdeplot(data=iris_data, x='Petal Width (cm)', fill=True, alpha=0.5, label='Largeur Pétale')
    sns.kdeplot(data=iris_data, x='Sepal Width (cm)', fill=True, alpha=0.5, label='Largeur Sépale')
    
    plt.title("Distribution des dimensions des Sépales et Pétales", fontsize=16, pad=15)
    plt.xlabel("Dimensions (cm)", fontsize=12)
    plt.ylabel("Densité", fontsize=12)
    plt.legend()
    plt.show()

    # ==========================================
    # 2. RELATIONS BIVARIÉES (Jointplots)
    # ==========================================
    
    # Focus sur les Pétales
    jp_petal = sns.jointplot(data=iris_data, x='Petal Length (cm)', y='Petal Width (cm)', hue='Species', height=8)
    jp_petal.fig.suptitle("Relation : Longueur vs Largeur des Pétales", y=1.02, fontsize=14)
    plt.show()

    # Focus sur les Sépales
    jp_sepal = sns.jointplot(data=iris_data, x='Sepal Length (cm)', y='Sepal Width (cm)', hue='Species', height=8)
    jp_sepal.fig.suptitle("Relation : Longueur vs Largeur des Sépales", y=1.02, fontsize=14)
    plt.show()

    # ==========================================
    # 3. RÉSUMÉ STATISTIQUE (Boxplot)
    # ==========================================
    plt.figure(figsize=(12, 6))
    
    sns.boxplot(data=iris_data, x='Species', y='Petal Length (cm)', hue='Species', legend=False, palette="Set2")
    
    plt.title("Résumé Statistique : Longueur des pétales par Espèce", fontsize=16, pad=15)
    plt.xlabel("Espèce d'Iris", fontsize=12)
    plt.ylabel("Longueur des Pétales (cm)", fontsize=12)
    plt.show()

else:
    print("❌ Erreur : Impossible de trouver le fichier CSV. Vérifie l'arborescence du projet.")
```
