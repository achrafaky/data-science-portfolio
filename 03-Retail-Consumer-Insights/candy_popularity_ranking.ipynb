```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Configuration du style
sns.set_theme(style="darkgrid")

# Chemin intelligent pour candy.csv
if os.path.exists("03-Retail-Consumer-Insights/candy.csv"):
    file_path = "03-Retail-Consumer-Insights/candy.csv"
else:
    file_path = "candy.csv"

if os.path.exists(file_path):
    print(f"⏳ Chargement des données depuis : {file_path} ...")
    candy_data = pd.read_csv(file_path, index_col="id")
    print("✅ Fichier chargé avec succès !")

    # -------------------------------------------------------
    # GRAPHIQUE 1 : SCATTER PLOT
    # -------------------------------------------------------
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=candy_data, x="sugarpercent", y="winpercent", hue="chocolate", palette="Set1", s=100, alpha=0.7)
    plt.title("Relation entre le sucre et les gagnants", fontsize=14)
    plt.xlabel("Pourcentage de sucre", fontsize=12)
    plt.ylabel("Pourcentage de gagnants", fontsize=12)
    plt.show()

    # -------------------------------------------------------
    # GRAPHIQUE 2 : LMPLOT (Régression)
    # -------------------------------------------------------
    sns.lmplot(data=candy_data, x="sugarpercent", y="winpercent", hue="chocolate", height=6, aspect=1.5)
    plt.title("Comparaison des pentes : Chocolat vs Pas Chocolat", fontsize=14)
    plt.show()

    # -------------------------------------------------------
    # GRAPHIQUE 3 : SWARMPLOT
    # -------------------------------------------------------
    plt.figure(figsize=(10, 6))
    sns.swarmplot(data=candy_data, x="chocolate", y="winpercent", hue="chocolate", palette="Set2", legend=False)
    plt.title("Distribution des gagnants selon la présence de chocolat", fontsize=14)
    plt.xlabel("Chocolat (0 = Non, 1 = Oui)", fontsize=12)
    plt.ylabel("Pourcentage de gagnants", fontsize=12)
    plt.show()

else:
    print(f"❌ Erreur : Impossible de trouver le fichier CSV. Vérifie son emplacement.")
```
